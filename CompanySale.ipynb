```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/matplotlib/company_sales_data.csv"
```

```python
df = pd.read_csv(url)

print(df.head())
```

```python
months = df['month_number']
total_profit = df['total_profit']

plt.plot(months, total_profit, marker='o')

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')

plt.show()
```

```python
months = df['month_number']
total_profit = df['total_profit']

plt.plot(months, total_profit, color='green', marker='o', linestyle='--', linewidth=2, markersize=6)

plt.xlabel('Month Number')
plt.ylabel('Total Profit')
plt.title('Total Profit by Month')

plt.show()
```

```python
products = df.columns[1:-2]
sales_data = df.iloc[:, 1:-2] 

for product in products:
    plt.plot(df['month_number'], sales_data[product], marker='o', label=product)

plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Product Sales by Month')
plt.legend()

plt.show()
```

```python
months = df['month_number']
toothpaste_sales = df['toothpaste']

plt.scatter(months, toothpaste_sales)

plt.xlabel('Month Number')
plt.ylabel('Toothpaste Sales')
plt.title('Toothpaste Sales by Month')

plt.show()
```

```python
products = ['facecream', 'facewash']
sales_data = df[products]

plt.bar(df['month_number'], sales_data['facecream'], label='Face Cream')
plt.bar(df['month_number'], sales_data['facewash'], label='Face Wash')

plt.xlabel('Month Number')
plt.ylabel('Sales')
plt.title('Face Cream and Face Wash Sales by Month')
plt.legend()

plt.show()
```

```python
months = df['month_number']
bathing_soap_sales = df['bathingsoap']

plt.bar(months, bathing_soap_sales)

plt.xlabel('Month Number')
plt.ylabel('Bathing Soap Sales')
plt.title('Bathing Soap Sales by Month')

plt.savefig('bathing_soap_sales_plot.png')

plt.show()
```

```python
total_profit = df['total_profit']

plt.hist(total_profit, bins=10)

plt.xlabel('Total Profit')
plt.ylabel('Frequency')
plt.title('Total Profit Distribution')

plt.show()
```

```python
product_sales = df.iloc[:, 1:7].sum()

labels = product_sales.index
sizes = product_sales.values
colors = ['blue', 'orange', 'green', 'red', 'purple', 'brown']
explode = (0.1, 0, 0, 0, 0, 0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.title('Total Sales Data for Last Year by Product')

plt.show()
```

```python
data = df[['month_number', 'bathingsoap', 'facewash']]

fig, axes = plt.subplots(nrows=2, ncols=1, figsize=(8, 10))

axes[0].plot(data['month_number'], data['bathingsoap'], marker='o', linestyle='-', color='blue')
axes[0].set_xlabel('Month')
axes[0].set_ylabel('Bathing soap')
axes[0].set_title('Bathing soap Sales')

axes[1].plot(data['month_number'], data['facewash'], marker='o', linestyle='-', color='green')
axes[1].set_xlabel('Month')
axes[1].set_ylabel('Facewash')
axes[1].set_title('Facewash Sales')

plt.tight_layout()

plt.show()
```

```python
data = df[['month_number', 'facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']]

months = data['month_number']

facecream_sales = data['facecream']
facewash_sales = data['facewash']
toothpaste_sales = data['toothpaste']
bathingsoap_sales = data['bathingsoap']
shampoo_sales = data['shampoo']
moisturizer_sales = data['moisturizer']

plt.stackplot(months, facecream_sales, facewash_sales, toothpaste_sales, bathingsoap_sales, shampoo_sales, moisturizer_sales, labels=['Face Cream', 'Facewash', 'Toothpaste', 'Bathing Soap', 'Shampoo', 'Moisturizer'], colors=['blue', 'orange', 'green', 'red', 'purple', 'brown'])
plt.xlabel('Month')
plt.ylabel('Sales')
plt.title('Product Sales Data')
plt.legend(loc='upper left')

plt.show()
```
