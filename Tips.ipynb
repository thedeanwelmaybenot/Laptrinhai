```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/tips.csv"
```

```python
tip = pd.read_csv(url)

print(tip.head())
```

```python
tip = tip.drop('Unnamed: 0', axis=1)

print(tip.head())
```

```python
plt.hist(tip['total_bill'], bins=10, edgecolor='black')
plt.xlabel('Total Bill')
plt.ylabel('Frequency')
plt.title('Historgram of Total Bill')
plt.show()
```

```python
plt.scatter(tip['total_bill'], tip['tip'])
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot: Total Bill vs Tip')
plt.show()
```

```python
plt.scatter(tip['total_bill'], tip['tip'], s=tip['size']*10, alpha=0.7)
plt.xlabel('Total Bill')
plt.ylabel('Tip')
plt.title('Scatter Plot: Total Bill, Tip and Size')
plt.show()
```

```python
sns.barplot(x='day', y='total_bill', data=tip)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Average Total Bill by day')
plt.show()
```

```python
plt.scatter(tip['tip'], tip['day'], c=tip['sex'].map({'Male': 'blue', 'Female': 'red'}))
plt.xlabel('Tip')
plt.ylabel('Day')
plt.title('Scatter Plot: Tip vs Day (Differentiated by Sex)')
plt.show()
```

```python
sns.boxplot(x='day', y='total_bill', hue='time', data=tip)
plt.xlabel('Day')
plt.ylabel('Total Bill')
plt.title('Scatter Plot: Tip vs Day (Differntiated bt Time)')
plt.legend(title='Time', loc='upper right')
plt.show()
```

```python
dinner_tip = tip[tip['time'] == 'Dinner']
lunch_tip = tip[tip['time'] == 'Lunch']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].hist(dinner_tip['tip'], bins=10, edgecolor='black')
axes[0].set_title('Dinner Tip')
axes[0].set_xlabel('Tip')
axes[0].set_ylabel('Frequency')

axes[1].hist(lunch_tip['tip'], bins=10, edgecolor='black')
axes[1].set_title('Lunch Tip')
axes[1].set_xlabel('Tip')
axes[1].set_ylabel('Frequency')

plt.tight_layout()

plt.show()
```

```python
male_tip = tip[tip['sex'] == 'Male']
female_tip = tip[tip['sex'] == 'Female']

fig, axes = plt.subplots(1, 2, figsize=(10, 4))

axes[0].scatter(male_tip['total_bill'], male_tip['tip'], c=male_tip['smoker'].map({'Yes': 'red', 'No': 'blue'}))
axes[0].set_title('Male Tip')
axes[0].set_xlabel('Total Tip')
axes[0].set_ylabel('Tip')

axes[1].scatter(female_tip['total_bill'], female_tip['tip'], c=female_tip['smoker'].map({'Yes': 'red', 'No': 'blue'}))
axes[1].set_title('Female Tip')
axes[1].set_xlabel('Total Bill')
axes[1].set_ylabel('Tip')

legend_labels = ['Non-Smoker', 'Smoker']
colors = ['blue', 'red']
legend_handles = [plt.Line2D([], [], marker='o', color=color, linestyle='', label=label) for color, label in zip(colors, legend_labels)]
plt.legend(handles=legend_handles, loc='upper right')

plt.tight_layout()

plt.show()
```
