```python
import pandas as pd
import matplotlib.pyplot as plt
import warnings
from collections import Counter

# set this so the graphs open internally
%matplotlib inline
```

```python
url = "https://raw.githubusercontent.com/thieu1995/csv-files/main/data/pandas/titanic_train.csv"
```

```python
titanicDesaster = pd.read_csv(url)

print(titanicDesaster.head())
```

```python
titanicDesaster.set_index('PassengerId', inplace=True)

print(titanicDesaster.head())
```

```python
male_count = titanicDesaster['Sex'].value_counts()['male']
female_count = titanicDesaster['Sex'].value_counts()['female']

labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['blue', 'pink']
explode = (0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.title('Male/Female Proportion')

plt.show()
```

```python
male_count = titanicDesaster['Sex'].value_counts()['male']
female_count = titanicDesaster['Sex'].value_counts()['female']

labels = ['Male', 'Female']
sizes = [male_count, female_count]
colors = ['blue', 'pink']
explode = (0, 0.1)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal') 
plt.title('Male/Female Proportion')

plt.show()
```

```python
plt.scatter(titanicDesaster['Fare'], titanicDesaster['Age'], c=titanicDesaster['Sex'].map({'male': 'blue', 'female': 'pink'}), alpha=0.5)

plt.xlabel('Fare')
plt.ylabel('Age')
plt.title('Fare vs Age')

legend_labels = ['Male', 'Female']
colors = ['blue', 'pink']
legend_handles = [plt.Line2D([], [], marker='o', color=color, linestyle='', label=label) for color, label in zip(colors, legend_labels)]
plt.legend(handles=legend_handles, loc='upper right')

plt.show()
```

```python
survived_count = titanicDesaster['Survived'].sum()

print("Count of individuals who survived:", survived_count)
```

```python
plt.hist(titanicDesaster['Fare'], bins=10, edgecolor='black')

plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Histogram of Fare Paid')

plt.show()
```
